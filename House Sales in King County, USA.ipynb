```python
data['sqft_above'] = data['sqft_above'].fillna(data['sqft_above'].mean())
data.isna().sum()
```

```python
x_train2 = data[["floors","waterfront","lat","bedrooms","sqft_basement","view","bathrooms","sqft_living15","sqft_above","grade","sqft_living"]]
y_train2 = data['price']
lr.fit(x_train2, y_train2)
lr.score(x_train2,y_train2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

Input = [('scale', StandardScaler()),("polynomial", PolynomialFeatures(degree =3)),('model', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(x_train2, y_train2)
pipe.score(x_train2,y_train2)
```

```python
from sklearn.linear_model import Ridge
RR = Ridge(alpha = 0.1)
RR.fit(x_train2, y_train2)
RR.score(x_train2, y_train2)
```

```python
from sklearn.model_selection import train_test_split
x_train,  x_test, y_train, y_test = train_test_split( x_train2, data['price'], test_size=0.33, random_state=42)
```

```python
pf = PolynomialFeatures(degree=2)
x_train_pf = pf.fit_transform(x_train)
x_test_pf = pf.fit_transform(x_test)
```

```python
from sklearn.linear_model import Ridge
rreg = Ridge(alpha =0.1)
rreg.fit(x_train_pf, y_train)
rreg.score(x_test_pf, y_test)
```
